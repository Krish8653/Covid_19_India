```python
import simplejson as json
import pandas as pd
```

```python
with open('Mohfw.json', 'r') as f:
    json_text = f.read()
```

```python
covid = json.loads(json_text)
```

```python
del covid['offset']
```

```python
rows = []
list = covid['rows']
```

```python
list
```

```python
new_list = []
for d in list:
    new_list.append(d['value'])
```

```python
new_list
```

```python
df = pd.DataFrame(new_list)
```

```python
df.head()
```

```python
clms = ['_id', '_rev']
```

```python
df.drop(clms, axis = 1, inplace = True)
```

```python
df['date'] = df['report_time'].str[:10]
```

```python
df['date'] = pd.to_datetime(df['date'], errors='coerce')
```

```python
df.drop('report_time', axis = 1, inplace = True)
```

```python
needed_data = df[(df['date'] > '2020-10-07') & (df['date'] < '2020-10-20')].copy()
```

```python
needed_data.drop('index', axis = 0, inplace = True)
```
