```python
import simplejson as json
import pandas as pd
```

```python
with open('Mohfw data.json', 'r') as f:
    json_text = f.read()
```

```python
covid = json.loads(json_text)
```

```python
del covid['offset']
```

```python
rows = []
list = covid['rows']
```

```python
list
```

```python
new_list = []
for d in list:
    new_list.append(d['value'])
```

```python
new_list
```

```python
df = pd.DataFrame(new_list)
```

```python
df.head()
```

```python
clms = ['_id', '_rev']
```

```python
df.drop(clms, axis = 1, inplace = True)
```
